```python
import seaborn as sns 
sns.histplot(titanic_copy["Fare"],kde=True)
```

```python
#Number of passengers who paid fare of 50 pecenrile 
percentile_50_fare = np.percentile(titanic_copy["Fare"],[50 ])

for i in range(len(titanic_copy["Fare"])):
    if titanic_copy["Fare"][i] >= percentile_50_fare:
        print(f'{titanic_copy["Name"][i]} paid fare of ==> {titanic_copy["Fare"][i]}')
    
```

```python
# Calculate percentile of age of each passenger
from scipy.stats import percentileofscore

# Loop through each row using range(len())
for i in range(len(titanic_copy)):
    age = titanic_copy["Age"].iloc[i]  # Use .iloc to access row by index
    
    # Check if Age is not null
    if pd.notnull(age):
        percentile = percentileofscore(titanic_copy["Age"].dropna(), age)
        print(f'{titanic_copy["Name"].iloc[i]} is {percentile:.2f}')
```

```python
titanic_copy["Age_Percentile"] = titanic_copy["Age"].apply(
    lambda x: percentileofscore(titanic_copy["Age"].dropna(), x) if pd.notnull(x) else np.nan
)


```

```python
for i in range(len(titanic_copy.columns)):  # Loop through columns
    column = titanic_copy.columns[i]        # Get column name
    
    if titanic_copy[column].isnull().sum() > 0:
        print(f'Column {column} contains null values')
    else:
        print(f'{column} does not contain null values')


```

```python
#Created four quaters of percentile for fare colmn of titnic data set
q1,q2,q3,q4 = np.percentile(titanic_copy["Fare"],[25,50,75,100])

IQR = q3 - q1
IQR

lower_fence= q1-(IQR*1.5)
upper_fence= q3+(IQR*1.5)

print(f'{lower_fence}, \n{upper_fence}')

import seaborn as sns
sns.boxplot(titanic_copy["Fare"])
```

```python
for i in range(len(titanic_copy)):
    if pd.notnull(titanic_copy["Fare"][i]) and titanic_copy["Fare"][i] > upper_fence:
        print(f'Name: {titanic_copy["Name"].iloc[i]}, Fare: {titanic_copy["Fare"].iloc[i]}')

```

```python
outlier = []

# Loop through each row
for i in range(len(titanic_copy)):
    if pd.notnull(titanic_copy["Fare"][i]) and titanic_copy["Fare"][i] > upper_fence:
        outlier.append(titanic_copy["Fare"][i])  # Append outliers to list
        print(f'{titanic_copy["Name"][i]} | Fare: {titanic_copy["Fare"][i]}')

print(f'\nTotal Number of Outliers: {len(outlier)}')



```
