```python
import pandas as pd
import matplotlib.pyplot as plt

# Corrected file path
titanic = pd.read_csv(r"E:\Dataset\Titanoc Dataset\Titanic-Dataset.csv")

# Convert 'Sex' to numeric values (0 = female, 1 = male)
titanic["Sex_Num"] = titanic["Sex"].map({"male": 1, "female": 0})

# Scatter plot (Better for Age vs Sex visualization)
plt.figure(figsize=(8, 5))
plt.scatter(titanic["Age"], titanic["Sex_Num"], c=titanic["Sex_Num"], cmap="coolwarm", alpha=0.5)
plt.xlabel("Age")
plt.ylabel("Sex (0 = Female, 1 = Male)")
plt.title("Titanic: Age vs Sex")
plt.grid(True)
plt.show()
```

```python
#Bar plot

np.random.seed(1)
x1= np.random.randint(0,15,5)
y1=np.random.randint(13,30,5)
x2=np.random.randint(-10,-3,5)
y2=np.random.randint(-20,-11,5)

print(f"Original Matrix is: \n {[x1]} : \n {[y1]}: \n {[x2]}: \n {[y2]}")
plt.bar(x1,y1,width=0.4)
plt.bar(x2,y2,width=0.4)
plt.xlabel('X Axis') 
plt.ylabel('Y Axis ')
plt.axhline(0, color='black', linewidth=1.2, linestyle="--")
print(np.max([y1, y2,x1,x2]))

```
