Example of Normal Distributed data

```python
normal1 = np.random.normal(0.5,5,1000)  #np.random.normal(mu,sigma,number of datapoints)
normal1
```

```python
sns.histplot(normal1,kde=True)
```

```python
mu = 0.8
sigma= 0.6
log_normal1 = np.random.lognormal(mu,sigma,100)
log_normal1
```

```python
sns.histplot(log_normal1)
```

```python
log_of_lognormal = np.log(log_normal1)
sns.histplot(log_of_lognormal,kde=True)
```

```python
import scipy.stats as stat
import pylab

def plot_data(data):
    plt.figure(figsize=(10,6),facecolor='orange')     #Figure allow you to control size of backgroung, colour of background etc
    plt.subplot(3,3,1)                                 #subplot allow you to showcase muliple plot subplot(row,column,position of plot)
    sns.histplot(data)
    plt.subplot(3,3,3)
    stat.probplot(data,dist="norm",plot=pylab)
    plt.tight_layout()
    plt.show()
```

```python
plot_data(log_normal1)
```

```python
plot_data(log_of_lognormal)
```

```python
#Bernoulies distribution
seed = 1
sample_size = 300
p= 0.7
np.random.seed(seed)
berno = np.random.choice([0,1],sample_size,p=[1-p,p])
berno
```

```python
sns.histplot(berno,bins=2,kde=False)
```

```python
# Binomial Distribution data
no_tiral = 10
probability = 0.6
sample_size = 800
seed = 1
np.random.seed(seed)
binomial1= np.random.binomial(no_tiral,probability,sample_size)
```

```python
sns.histplot(binomial1)
```

```python
#Poissoms distribution 
lam = 5
sample_size = 500
poisson1 = np.random.poisson(lam,sample_size)
poisson1
```

```python
sns.histplot(poisson1)
```
